```python
def find_orf(seq):
    # identify the longest open reading frame (orf) in any reading frame for seq
    
    stop_codons = ['TAA', 'TAG', 'TGA']
    
    # split into 6 reading frames
    # forward
    frame_1 = [seq[i:i+3] for i in range(0, len(seq), 3)]
    frame_2 = [seq[i:i+3] for i in range(1, len(seq), 3)]
    frame_3 = [seq[i:i+3] for i in range(2, len(seq), 3)]
    
    # reverse
    rev_seq = seq[::-1]
    frame_4 = [rev_seq[i:i+3] for i in range(0, len(rev_seq), 3)]
    frame_5 = [rev_seq[i:i+3] for i in range(1, len(rev_seq), 3)]
    frame_6 = [rev_seq[i:i+3] for i in range(2, len(rev_seq), 3)]
    
    print('frame_1: ',frame_1[:3],'frame_2: ',frame_2[:3],'frame_3: ',frame_3[:3])
    
    frames = [frame_1, frame_2, frame_3]
    
    longest_orf = 0
    
    for j in range(len(frames)):
        
        i = 0
        start = 0
        orf_state = False

        for codon in frames[j]:
            if codon == 'ATG' and orf_state == False:
#                 print('start codon found: ', i)
                start = i
                orf_state = True
            if codon in stop_codons and orf_state == True:
                
                orf = frames[j][start:i+3]
                orf = ''.join(orf)
                
                if len(orf) > longest_orf:
                    print('len of orf: ', len(orf))
                    longest_orf = len(orf)
                    print('stop codon {} found: {}'.format(codon, i))
                    print('Longest orf ' + str(len(orf)) + ' nt @ pos : ' + str(start) + ' in frame {}'.format(j+1))
                
                    print(orf)
                orf_state = False
                
            i += 1
    print('Longest orf :', str(longest_orf))
```

```python
# Find the most frequently occurring repeat of length 6 in all sequences. How many times does it occur in all?


repeat_6 = {}

for read in reads[1:]:
    seq = (read.split('\n')[1:])
    seq = ''.join(seq)
    print(seq[:10])
    len_seq = len(seq)
    print(len(seq))
    
    # forward
    windows = [seq[i:i+6] for i in range(len(seq))]
    for window in windows:
        if window in repeat_6:
            repeat_6[window] += 1
        else:
            repeat_6[window] = 1
    
    # reverse
    rev_seq = seq[::-1]
    windows = [rev_seq[i:i+6] for i in range(len(rev_seq))]
    for window in windows:
        if window in repeat_6:
            repeat_6[window] += 1
        else:
            repeat_6[window] = 1

print(len(repeat_6))
```
